# Matplotlib

- Matplotlib es una biblioteca para la generación de gráficos a partir de datos contenidos en listas o arrays en un lenguaje de programación Python.
- Matplotlib te da el control sobre cada aspecto del gráfico.
- Si tienes instalado la distribución Anaconda, no deberias de tener ningun problema para usarlo, al instalar adecuadamente anaconda.
- De no ser el caso, puedes instalar desde el terminal mediante la ejecución del siguiente comando: pip install matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(2,9,10)
x
```

```python
y = x**2
```

```python
y
```

```python
#Graficar una grafica con nuestros datos x - y
plt.plot(x,y,'red')
plt.title('Este es mi grafico')
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.show()
```

```python
#varios graficos en el mismo espacio con funcion subplot
plt.subplot(1,2,1)
plt.plot(x,y,'g')

plt.subplot(1,2,2)
plt.plot(y,x,'blue')
```

- Programacion orientada a objetos

```python
#Creacion de graficos dentro de una misma figura
figura = plt.figure()
grafico = figura.add_axes([0.1,0.1,0.8,0.8])#
grafico.plot(x,y,'r')

grafico1 = figura.add_axes([0.15,0.5,0.3,0.3])#[x,y,ancho, alto] con y-x ,me refiero al porcentaje n ecual se encuentran en esa
                                                #cordenada en y o arriba o abajo, en x derecha o izquierda
grafico1.plot(y,x, 'blue')
```

# Multigráficos (filas, columnas)

```python
#Fucniona como una matriz, filas y columnas
figura, graficos = plt.subplots(nrows=2, ncols=2)

#Grafico de la curava y su color
graficos[0][0].plot(x,y, 'b')
graficos[0][1].plot(x,y, 'r')
graficos[1][0].plot(x,y, 'y')
graficos[1][1].plot(x,y, 'g')

#Nombres de los graficos
graficos[0][0].set_title('Azul')
graficos[0][1].set_title('Rojo')
graficos[1][0].set_title('Amarillo')
graficos[1][1].set_title('Verde')
```

## Tamaño del gráfico

```python
figura = plt.figure(figsize=(8,4))# Aqui se realiza la configuracion del tamaño,(ancho, alto)
grafico = figura.add_axes([0.1, 0.1, 0.9, 0.9])
grafico.plot(x, y, 'r')
```

```python
figura = plt.figure(figsize=(5,8))# Aqui se realiza la configuracion del tamaño,(ancho, alto)
grafico = figura.add_axes([0.1, 0.1, 0.9, 0.9])
grafico.plot(x, y, 'b')
```

# Dos gráficos en la misma figura

```python
figura = plt.figure()
graficos = figura.add_axes([0.1, 0.1, 0.9, 0.9])
graficos.plot(x, x**2, label = 'Al cuadrado')
graficos.plot(x, x**3, label = 'Al cubo')

#Añadir una leyenda para diferenciar una de otra
graficos.legend(loc=0)
```

## Color de Grárfico

```python
figura = plt.figure()
grafico = figura.add_axes([0,0,1,1])
grafico.plot(x,y, 'lime', label = 'documentacion')# lsita de colores aqui https://matplotlib.org/stable/gallery/color/named_colors.html
# y tambien se puede elegir color por codigos hexagesimales
grafico.plot(x,x**3, '#a4ccee', label = 'hexagesimal')
grafico.legend(loc=0)
```

## Tipos de lineas del gráfico(grosor, transparencia y tipo)

```python
figura = plt.figure()
grafico = figura.add_axes([0,0,1,1])
#linewidth =  Ancho de linea    alpha = Transparencia de linea linestyle = tipo de linea
grafico.plot(x,y,color = 'red', linewidth = 10, alpha = 0.4, linestyle = '-')

#tipos de linea en la documentacion: https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html
```
